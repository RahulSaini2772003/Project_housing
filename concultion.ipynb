```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hs = pd.read_csv('housing.csv')
hs
```

```python
plt.figure(figsize=(16,9),dpi=200)
plt.scatter(hs['longitude'],hs['latitude'])
plt.xlabel('Latitude',size=20)
plt.ylabel('longitude',size=20)
```

```python
hs['ocean_proximity'].unique()
```

```python
color = hs['ocean_proximity'].map({
    'NEAR BAY' : 'y',
    '<1H OCEAN' : 'c',
    'INLAND' : 'r',
    'NEAR OCEAN' : 'g',
    'ISLAND' :'b'
})
```

```python
plt.figure(figsize=(16,9),dpi=200)
plt.scatter(hs['longitude'],hs['latitude'],color = color)
plt.xlabel('Latitude',size=20)
plt.ylabel('longitude',size=20)
plt.show()
```

```python
ind = hs['ocean_proximity'].value_counts().index
va = hs['ocean_proximity'].value_counts().values
```

```python
plt.pie(va,labels=ind,explode=[.1,.1,.1,.1,.1], shadow=True)
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
i1 =  tips['smoker'].value_counts().index
v1 = tips['smoker'].value_counts().values
i2 =  tips.groupby('size')['total_bill'].sum().index
v2 = tips.groupby('size')['total_bill'].sum().values
```

```python
fig = plt.figure(figsize=(16,9),dpi=200,facecolor='k')
ax1 = fig.add_axes([0,0,0.5,0.5])
ax2 = fig.add_axes([0,0.6,0.5,0.5])
ax3 = fig.add_axes([0.6,0.6,0.5,0.5])
ax4 = fig.add_axes([0.6,0,0.5,0.5])

ax1  = plt.scatter(tips['total_bill'],tips['tip'])
ax2.bar(i1,v1, color= ['r','b'])
ax3.bar(i2,v2, color= [*'rbgy'])
ax4.hist(tips['total_bill'],bins = 50, ec = 'r')

for ax in (ax1,ax2,ax3,ax4):
    ax.set_fc('k')
    ax.tick_params(color = 'w')
```
